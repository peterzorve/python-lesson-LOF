```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python

def f(x):
    return np.sin(x)

def g(x):
    return np.cos(x)
```

```python
x = np.linspace(-5, 5, 25)
# x = range(-5, 5, 1)

y = f(x)
t = g(x)

plt.plot(x, y)
plt.plot(x, t)
# print(x)
```
